```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn

from sklearn.datasets import fetch_california_housing

%matplotlib inline
```

```python
data = fetch_california_housing()
```

```python
display(data)
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
display(df.head(5)) # Displaying first 5 records of the dataset
display(df.tail(5)) # Display last 5 records of the dataset
```

```python
df['MEDV'] = data.target
display(df.head(5)) # Displaying first 5 records of the dataset
display(df.tail(5)) # Display last 5 records of the dataset
```

```python
print(df.dtypes)
```
